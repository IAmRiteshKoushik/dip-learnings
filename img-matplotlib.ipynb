```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
image = cv.imread("Photos/cat.jpg")
plt.imshow(image)
plt.show()

# Jupyter crashes if opencv is directly used to spawn a windowThread and 
# display the image. An alternative is to render it on the client side using 
# matplotlib's pyplot module which has an imshow() function. 

# The problem is that this renders the image in BGR.
```

```python
image = cv.imread("Photos/cat.jpg")
im2 = image.copy()
im2[:, :, 0] = image[:, :, 2]
im2[:, :, 2] = image[:, :, 0]
plt.imshow(im2)
plt.show()
```

```python
# Showing them side by side
image = cv.imread("Photos/cat.jpg")
image2 = image.copy()

fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.imshow(image)
# Default matplotlib tends to blue-shade images as it 
# reads images in a BGR mode instead of RGB. Need explicit
# conversion
ax1.set_title("Matplotlib Image (BGR) ")
ax1.axis("off")

ax2.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
ax2.set_title("OpenCV Image(RB")
ax2.axis("off")

plt.tight_layout()
plt.show()
```
